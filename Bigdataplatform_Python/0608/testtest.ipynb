```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
# 한글문제 해결처리 
################################################################
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
################################################################
df = pd.read_csv("21.모범음식점현황(사전정보공표181231).csv", encoding='cp949') 

#1번 모범음식점현황(건수)을 그래프로 출력하시오.(업태기준)
df2 =df.groupby(by='업태').size().reset_index(name='건수')
plt.plot(df2['업태'],df2['건수'],'g')
plt.title("모범음식점 현황")
plt.xlabel("업태")
plt.ylabel("건수")
plt.show()

#2번 “거창군”과 “함양군”에서 (메뉴 등장 빈도수가) 하위인 4곳의 음식점을 찾고 4곳의 음식점 기준으로 그래프를 출력하시오.
filter1 = df['시군구'] =='거창군'
filter2 = df['시군구'] =='함양'
sigun1 = df[filter1]
sigun2 = df[filter2]

findmenu = pd.concat([sigun1, sigun2])
findmenu

df3 = findmenu.groupby(by='주메뉴').size().reset_index(name='건수')
df4 = df3.sort_values(by='건수', ascending=False).tail(4)
df4
plt.plot(df4['주메뉴'],df4['건수'],'g')
plt.show()

#3번 “낙지”를 포함하는 주메뉴의 상위 3건을 찾고 그에 해당하는 그래프를 출력하시오.(주메뉴기준)
df5 = df[df['주메뉴'].str.match('낙지')]
df6 = df5.groupby(by='주메뉴').size().reset_index(name='건수')
df7 = df6.sort_values(by='건수', ascending=False).head(3)
df7
plt.plot(df7['주메뉴'],df7['건수'],'g')
plt.show()
```
