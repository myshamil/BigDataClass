```python
import json
import requests
import numpy as np
import pandas as pd
import folium
import xmltodict
# !pip install xmltodict

request_url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19SidoInfStateJson?serviceKey=gZ1mis9Xa%2FvXgaEyc0FxYyZ7QvEq1u5WTu4U5thsna%2Bbp3lNV0RztdYSw%2FIuvOIZkJWjETFZRUMG%2FcpQuJgHwA%3D%3D&pageNo=1&numOfRows=10&startCreateDt=20200410&endCreateDt=20200410'
res = requests.get(request_url)
xpars = xmltodict.parse(res.text)
json_str = json.dumps(xpars)
json_data = json.loads(json_str, encoding='euc-kr')
json_items = json_data['response']['body']['items']['item']

data = []
for item in json_items:
    # print(item['gubun'], item['qurRate'])
    gubun, qurRate = item['gubun'], item['qurRate']
    data.append([gubun, qurRate])

df = pd.DataFrame(data, columns=['',''])
df2 = df.iloc[1:18,:]
df3 = df2.set_index('')
df3[''] = df3[''].astype(float)
df3

# geo_path = "20200610/skorea_municipalities_geo_simple.json"
geo_path = "20200610/05.skorea_municipalities_geo_simple.json"
geo_str = json.load(open(geo_path, encoding='utf-8'))
map = folium.Map(location=[37.566345, 126.977893])
map.choropleth(geo_data = geo_str, data=df3[''], columns=[df3.index, df3['']], fill_color='PuRd', )
map

```
