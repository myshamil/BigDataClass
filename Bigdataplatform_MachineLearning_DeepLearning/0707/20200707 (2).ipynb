```python
#범주형 -> 바차트
def barChart(f):
    survived=train[train['Survived']==1][f].value_counts()
    dead=train[train['Survived']==0][f].value_counts()
    df=pd.DataFrame([survived, dead])
    df.index=['Survived', 'Dead']
    df.plot(kind='bar', figsize=(10,5), stacked=True)   
```

```python
train=train.drop(['Ticket', 'SibSp', 'Parch'], axis=1)
test=test.drop(['Ticket', 'SibSp', 'Parch'], axis=1)                  
```

```python
train=train.drop(['PassengerId'], axis=1)
trainData=train.drop(['Survived'],axis=1)
target=train['Survived']
#trainData -> 모델 -> 예측  == 정답(target)
testData = test.drop("PassengerId", axis=1).copy()
```

```python
#CrossValidation
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import KFold, cross_val_score
import numpy as np
```

```python
# KFold 객체생성
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
```

```python
#KNN
clf = KNeighborsClassifier(n_neighbors=11)
score = cross_val_score(clf, trainData, target, cv=kfold,
                n_jobs=1, scoring='accuracy')
np.mean(score)
```

```python
#의사결정나무
clf = DecisionTreeClassifier()
score = cross_val_score(clf, trainData, target, cv=kfold,
                n_jobs=1, scoring='accuracy')
np.mean(score)
```

```python
#랜덤포레스트
clf = RandomForestClassifier(n_estimators=15) # n_estimators : 트리의 갯수
score = cross_val_score(clf, trainData, target, cv=kfold,
                n_jobs=1, scoring='accuracy')
np.mean(score)
```

```python
#베이즈
clf = GaussianNB() # n_estimators : 트리의 갯수
score = cross_val_score(clf, trainData, target, cv=kfold,
                n_jobs=1, scoring='accuracy')
np.mean(score)
```

```python
#SVM
clf = SVC() # n_estimators : 트리의 갯수
score = cross_val_score(clf, trainData, target, cv=kfold,
                n_jobs=1, scoring='accuracy')
np.mean(score)
```

```python
# 예측
clf.fit(trainData,target)
prediction = clf.predict(testData)
```
