```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import numpy as np
from keras.utils import to_categorical
```

```python
text="""들판에 있는 말이 뛰고 있다
그의 말이 법이다
가는 말이 고와야 오는 말이 곱다"""
```

```python
t=Tokenizer()
t.fit_on_texts([text])
```

```python
vocabSize=len(t.word_index)+1 #11+1 => 12
```

```python
#트레이닝 데이터 생성
sequences=[]
for line in text.split("\n"):
    encoded=t.texts_to_sequences([line])[0]
    for i in range(1,len(encoded)): #i= 1 to 4
        sequence=encoded[:i+1] #encoded[:3]
        sequences.append(sequence)
print(sequences)                
#         print(sequences)
#         print("="*50)
# [2, 3, 1, 4, 5]
# [6, 1, 7]
# [8, 1, 9, 10, 1, 11]
```

```python
# arr=[len(s) for s in sequences]
# max(arr)
maxLen=max(len(s) for s in sequences) #최대길이 6
```

```python
sequences=pad_sequences(sequences, maxlen=maxLen, padding='pre')
```

```python
x=sequences[:,:-1]
y=sequences[:,-1]
```

```python
y
```

```python
# ex) 1 => 0100000000000 ~ 11=>000000000001
y=to_categorical(y, num_classes=vocabSize)
y
```

```python
#모델 생성
from keras.models import Sequential
from keras.layers import Embedding, Dense, SimpleRNN, LSTM
```

```python
model=Sequential()
model.add(Embedding(vocabSize, 10, input_length=maxLen-1))
#입력 데이터의 길이는 maxlen에서 y를 뺀 maxLen-1이 됨
#임베딩 벡터의 차원은 10

model.add(SimpleRNN(32))
model.add(Dense(vocabSize, activation='softmax'))
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
model.fit(x,y,epochs=200)
```

```python
#'들판에' -> '들판에 있는 말이 뛰고 있다' 예측 완성

def sentence_generation(model, t, cw, n):
    initWord=cw
    sentence=''
    for _ in range(n):  #4번 반복
        encoded=t.texts_to_sequences([cw])[0] #[2]
        encoded=pad_sequences([encoded],maxlen=5, padding='pre')
        #print(encoded) #[2] -> [0 0 0 0 2]
        result=model.predict_classes(encoded)
        #print(result)   #[3] == '있는'
        for word, index in t.word_index.items():
            if index==result:
                break
        cw=cw+" "+word #"들판에 있는"
        sentence=sentence+" "+word
    sentence=initWord+sentence
    return sentence
    
sentence_generation(model, t, '가는', 2)
```
