#다시 원 데이터 불러오기

```python
#범주형 레코드를 원핫인코딩 
#sex-> sex_0 sex_1
# 0    1    0
#1     0    1
#Pclass -> Pclass1, Pclass2, Pclass3
#Embarked -> EnbarkedC, EmbarkedQ, EmbarkedS
```

```python
def myData(data,col):
    for c in col:
        data =pd.concat([data, pd.get_dummies(data[c], prefix=c)], axis=1)
        data =data.drop(c, axis=1)
    return data
```

```python
col=['Sex','Pclass','Embarked']
myData(train, col)
```
