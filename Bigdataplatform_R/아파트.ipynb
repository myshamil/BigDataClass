```python
# 한글문제 해결처리 
################################################################
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
################################################################
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pieces = []
for year in range(2012,2020):
    frame = pd.read_excel('데이터분석실습데이터/%d년.xls' %year , index_col=0)
    frame = frame.T
    pieces.append(frame)
df = pd.concat(pieces, ignore_index = True)
```

```python
df1 = df[['공용관리비','일반관리비','청소비','소독비','승강기유지비','지능형홈네트워크설비유지비','수선유지비','위탁관리수수료']]
df2 = df[['일반관리비','인건비','제사무비','제세공과금','피복비','교육훈련비','차량유지비','그밖의부대비용']]
```

```python
df1.corr()
```

```python
df2.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data = df1.corr(), annot=True, cmap='Blues')
```
