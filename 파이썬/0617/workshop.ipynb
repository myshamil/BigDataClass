```python
# 1. toji.txt  - 대화부분 (" ")추출 -> 명사 추출
# 2. 네이버 지식인  - 검색어 : 파이썬 =< 10페이지 질문/답변 추출 -> 명사 추출
# 3. 이메일 주소 패턴 정의 - park@google.com, df.park@test.co.kr ...등등 가능, kim@abc ->불가
import pandas as pd 
import numpy as np 
import re
import nltk
from konlpy.tag import Kkma
kkma= Kkma()

toji_text_raw = open('toji.txt', encoding='utf-16le').read()
saying =re.findall('[\"][가-힣\W가-힣]+[\"]', toji_text_raw)
nouns =[]
for i in saying:
    nouns.append(kkma.nouns(i))

nouns


```

```python
# 2. 네이버 지식인  - 검색어 : 파이썬 -> 10페이지 질문/답변 추출 -> 명사 추출
from bs4 import BeautifulSoup
from urllib.request import urlopen

for i in range(1,100,10):

    url = "https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&c_id=&c_name=&sm=tab_pge&kin_start=" + str(i)

    html = urlopen(url)
    soup = BeautifulSoup(html, "html.parser")
    qna = []
    for j in range(0,10):
        question = soup.findAll('dt','question')[j]
        answer = soup.findAll('dd','answer')[j].get_text()
        qna.append(question.find('a').get_text())
        qna.append(answer)

#명사 추출

```

```python
# 3. 이메일 주소 패턴 정의 - park@google.com, df.park@test.co.kr ...등등 가능, kim@abc ->불가
re.match("[A-Za-z\.]+[\@]+[A-Za-z\.]",'park@google.com')
re.match("[A-Za-z\.]+[\@]+[A-Za-z\.]",'kim@abc')
```
