```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
#주메뉴가 가장많은 상위 4건 추출하여 그래프 형태로 
# 한글문제 해결처리 
################################################################
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
################################################################
df = pd.read_csv("21.모범음식점현황(사전정보공표181231).csv", encoding='cp949') 
df

# df2= df['주메뉴']
# df3 = df2.str.split(',')
# df3

# df.describe()
df2 =df.groupby(by='주메뉴').size().reset_index(name='건수')
df3 = df2.sort_values(by='건수', ascending=False).head(4)
plt.bar(df3['주메뉴'],df3['건수'])

plt.show()

```
