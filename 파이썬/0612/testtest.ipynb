```python
import pandas as pd 
import matplotlib.pyplot as plt 
# 한글문제 해결처리 
################################################################
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
################################################################

df = pd.read_csv('/python_Pjt/경찰청_졸음운전교통사고_2014_2016.csv', encoding='cp949')
df.index
df.shape
df.columns
df.head()
df['구분'] = df['구분'].astype('str')
df['년도'] =df['구분'].str[:4]
df.head()

df1 =df.groupby(by="년도").sum()
df1.plot(kind = 'bar', rot = 0)
plt.title("년도별 사고횟수")
plt.xlabel("년도")
plt.ylabel("건수&명수")
plt.show()
```

```python
#2번, 3번
import nltk
from konlpy.tag import Twitter; t=Twitter()
from wordcloud import WordCloud, STOPWORDS
from collections import Counter

# 한글문제 해결처리 
################################################################
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
################################################################

text = open('키다리아저씨.txt').read()
nouns = t.nouns(text)
nouns
nouns = [n for n in nouns if len(n)>1]
count = Counter(nouns)

#tags = count.most_common(5)  #상위 5개 


tags = count.most_common(800)
tags

wc = WordCloud(background_color='white', max_words= 3000, font_path = "c:/Windows/Fonts/YBLA05.TTF")
cloud = wc.generate_from_frequencies(dict(tags))
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
