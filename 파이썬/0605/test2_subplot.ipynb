```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
fig = plt.figure()
x = np.arange(1,5)
a = fig.subplots(2,2)

a[0][0].grid(True) #grid
a[0][0].set_title("squere") #title
a[0][0].plot(x,x*x,'red',lw=10) #draw chart
a[0][1].plot(x,np.sqrt(x))
a[0][1].grid(color = 'blue', ls='-.', lw=3) # ls 선 종류, lw 선 두께
a[1][0].plot(x,np.exp(x))
a[1][1].plot(x,np.log10(x))
#plt.show()
```

```python
fig = plt.figure(figsize=(4,4)) #인치단위 설정
a1 = fig.add_axes([0,0,1,1])
x4 = np.arange(1,10)
a1.plot(x4, np.exp(x4))
a1.set_title('exp')
a1.set_ylim(0,10000) # y축 늘리기
a1.set_xlim(0,10) # x축 늘리기
```
