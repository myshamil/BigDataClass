```python
import numpy as np 
import pandas as pd 
import math
import matplotlib.pyplot as plt 
#시각화를 해봅시다

x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("SINE Graph")
plt.show()
```

```python
from pylab import *
x = linspace(-3, 3, 30)
y = x**2
y1 = x**4
plot(x,y,'rd')
plot(x,y1,'bp')
plot(x,y1+4,'m')
show()
```

```python
plot(x, sin(x),'p')
plot(x, cos(x),'r+')
plot(x, -sin(x),'m--')
```

```python

x1 = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x1)
fig = plt.figure() #기본 생성
ax = fig.add_axes([0,0,1,1]) #left,botton, width, height
ax.plot(x1,y1)
ax.legend(labels=('x'), loc = 'lower right')
ax.set_title('Sine graph')
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
plt.show()
```

```python
x3 = np.arange(0, math.pi*2,0.05)
y3 = sin(x3)
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.55,0.55,0.3,0.3])
axes1.plot(x3,y3,'b')
axes2.plot(x3, np.cos(y3), 'm')
axes1.set_title('sine')
axes2.set_title('cosine')
```
