```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

x=np.arange(1,5)
a = plt.subplot(2,2)
a[0][0].plot(x, x*x,'red', lw=10)
a[0][0].grid(True) #표 출력 유무
a[0][0].set_title("square")
a[0][1].plot(x, np.sqrt(x))
a[0][1].grid(color='blue', ls='-.', lw=1) #ls 선 종류, 두ㄲㅔ
a[0][1].set_title("square root")
a[1][0].plot(x, np.exp(x))
a[1][1].plot(x, np.log10(x))
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

x5 = np.arange(0, math.pi*2, 0.05)
y5 = np.sin(x5)
fig = plt.figure(figsize=(8,5))
a5 = fig.add_axes([0.1,0.1,0.8,0.8])
a5.plot(x5, np.exp(x5))
a5.set_ylabel('exp')
a6 =  a5.twinx()
a6.plot(x5, np.log(x5), 'm--')
a6.set_ylabel('log')
fig.legend(labels=('exp','log'), loc = 'upper left')

plt.show()
```

```python
#Bar Plot
fig= plt.figure(figsize=(9,4))
languages = ['java','javascript','python','html5','css3','sql','c++','c']
students = [29,15,20,3,2,10,15,20]
students2 = [25,10,10,19,10,15,1,15]

plt.bar(languages, students, color ='lightblue')
plt.bar(languages, students2, color ='c')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

data = [[30,25,50,20],[40,23,51,17],[35,22,45,19]]
x6 = np.arange(4)
fig = plt.figure(figsize=(8,4))
ax6 = fig.add_axes([0,0,1,1])
ax6.bar(x6 + 0.00, data[0], color='c', width = 0.25)
ax6.bar(x6 + 0.25, data[1], color='lightblue', width = 0.25)
ax6.bar(x6 + 0.50, data[2], color='gray', width = 0.25)
fig.legend(labels=('developer', 'dba','system engineer'), loc ="best")
plt.show()
```
